```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import zscore
```

```python
red = pd.read_csv('/content/winequality-red.csv', sep = ';')
red.describe()
#plt.scatter(red['fixed acidity'], red['quality'], )
red.corr()['quality']
red = red.apply(zscore)
```

```python
fig, ax = plt.subplots(nrows = 2, ncols = 2)
ax[0,0].scatter(red['alcohol'], red['quality'])

ax[0,1].scatter(red['volatile acidity'], red['quality'])
ax[1,1].scatter(red['sulphates'], red['quality'])
plt.show()
```
