```python
import pandas as pd
import numpy as np
import numpy.linalg as la

from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.linear_model import ElasticNet

from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

Load Wine Quality Data

```python
winequality_red = pd.read_csv('winequality-red.csv', sep=';')
winequality_white = pd.read_csv('winequality-white.csv', sep=';')
```

```python
RX, Ry = winequality_red.drop('quality', axis=1), winequality_red.quality
```

```python
RXtrain, RXtest, Rytrain, Rytest = train_test_split(RX, Ry)
```

```python
Rxtrain.head()
```

```python
Rxtrain.dtypes
```

```python
ROLS = LinearRegression().fit(RXtrain, Rytrain)
```

```python
Rypred = ROLS.predict(RXtest)
```

```python
mean_squared_error(Rytest, Rypred)
```

```python
mean_absolute_error(Rytest, Rypred)
```

```python
ROLS.score(RXtest, Rytest)
```

The Ordinary Least Squares Model did fairly well, on average the distance between the actual and predicted values was ~0.4. Less than half a point.

Now for the White Wine data:

```python
WX, Wy = winequality_white.drop('quality', axis=1), winequality_white.quality
```

```python
WXtrain, WXtest, Wytrain, Wytest = train_test_split(WX, Wy)
```

```python
WOLS = LinearRegression().fit(WXtrain, Wytrain)
```

```python
Wypred = WOLS.predict(WXtest)
```

```python
mean_squared_error(Wytest, Wypred)
```

```python
mean_absolute_error(Wytest, Wypred)
```

```python
WOLS.score(WXtest, Wytest)
```

This model did slightly worse as the MSE was approximatley 0.6.

Now let's try a ridge regression!

```python
#Note that alpha = 1.0 is the default
RRidge = Ridge().fit(RXtrain, Rytrain)
WRidge = Ridge().fit(WXtrain, Wytrain)
#I would love to do some research (potentially grid search) here to optimize this hyperparameter
```

```python
RypredR = RRidge.predict(RXtest)
WypredR = RRidge.predict(WXtest)
```

```python
print('Mean_Squared_Error_Red_Ridge:', mean_squared_error(Rytest, RypredR))
print('')
print('Mean_Squared_error_White_Ridge:', mean_squared_error(Wytest, WypredR))
```

```python
print('Mean_Squared_Error_Red_Ridge:', mean_squared_error(Rytest, RypredR))
print('')
print('Mean_Squared_error_White_Lasso:', mean_squared_error(Wytest, WypredR))
```

```python
print('RRidge score:', RRidge.score(RXtest, Rytest))
print('')
print('WRidge score:', WRidge.score(WXtest, Wytest))
```

The models here performed very similarly (and slightly worse) to the OLS models despite the l1 regularization.

Then we'll try Lasso regression with l2 regularization!

```python
RLasso = Lasso().fit(RXtrain, Rytrain)
WLasso = Lasso().fit(WXtrain, Wytrain)
```

```python
RypredL = RLasso.predict(RXtest)
WypredL = WLasso.predict(WXtest)
```

```python
print('Mean_Squared_Error_Red_Lasso:', mean_squared_error(Rytest, RypredL))
print('')
print('Mean_Squared_error_White_Lasso:', mean_squared_error(Wytest, WypredL))
```

```python
print('Mean_Absolute_Error_Red_Lasso:', mean_absolute_error(Rytest, RypredL))
print('')
print('Mean_Absolute_error_White_Lasso:', mean_absolute_error(Wytest, WypredL))
```

```python
print('RLasso score:', RLasso.score(RXtest, Rytest))
print('')
print('WLasso score:', WLasso.score(WXtest, Wytest))
```

This performed significantly worse than either of the other options so far! However, we still have yet to do any optimization of the parameters.

Finally: Let's do ElasticNet!

```python
RElasticNet = ElasticNet().fit(RXtrain, Rytrain)
WElasticNet = ElasticNet().fit(WXtrain, Wytrain)
```

```python
RypredEN = RElasticNet.predict(RXtest)
WypredEN = WElasticNet.predict(WXtest)
```

```python
print('Mean_Squared_Error_Red_ElasticNet:', mean_squared_error(Rytest, RypredEN))
print('')
print('Mean_Squared_error_White_ElasticNet:', mean_squared_error(Wytest, WypredEN))
```

```python
print('Mean_Absolute_Error_Red_ElasticNet:', mean_absolute_error(Rytest, RypredEN))
print('')
print('Mean_Absolute_error_White_ElasticNet:', mean_absolute_error(Wytest, WypredEN))
```

```python
print('RElasticNet score:', RElasticNet.score(RXtest, Rytest))
print('')
print('WElasticNet score:', WElasticNet.score(WXtest, Wytest))
```

This also performed significantly worse than the other models pre-optimization. I believe the next step will be to implement train_test_validate_split to try and optimize the hyperparameters of our models.
